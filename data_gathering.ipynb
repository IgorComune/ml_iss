# Knowing the Data

```python
#storing the json into a variable
iss = requests.get(link_api).json()
iss
```

```python
#selecting columns and values of the json
columns = iss.keys()
values = iss.values()

#returning column names
print('-'*120)
print(f'Columns names{columns}')
print('-'*120)
print(f'Number of columns: {len(columns)}')
print('-'*120)
```

```python
def iss_api():
    #storing the data and status of the API
    iss = requests.get('https://api.wheretheiss.at/v1/satellites/25544')
    resp = iss.status_code
    
    #retriving the JSON
    iss_api_data = iss.json()
    
    #checking API status
    if resp == 200:
        print ('API status: OK!')
    else:
        print ('De-orbiting WARNING!')
    
    return iss_api_data

#calling the iss_api function
iss_json = iss_api()
```

# JSON to Pandas

```python
#visualizing the JSON
iss_json
```

```python
# function to convert JSON to pandas dataframe
def json_df(json):
    iss_df = pd.json_normalize(json)
    return iss_df
```

# Appending new values

```python
# creating empty data frame
iss_main_df = pd.DataFrame(columns = columns)

for i in range(800):
    # getting data
    iss_json = iss_api()
    
    # transforming JSON to Pandas Data Frame
    new_row = json_df(iss_json)
    
    # appeding the row in the main Data Frame
    iss_main_df = iss_main_df.append(new_row)
    
    # getting data in invervals of X seconds
    sleep(30)
```

```python
# exporting the dataframe to csv file
iss_main_df.to_csv('iss.csv')

#reading the expoted CSV
pd.read_csv('iss.csv')
```
